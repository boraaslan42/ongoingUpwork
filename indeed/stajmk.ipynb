```python
from bs4 import BeautifulSoup
import re

def extract_emails_from_html(html):
    soup = BeautifulSoup(html, 'html.parser')
    # Extract text from HTML
    text = soup.get_text()
    # Use regular expression to find email addresses
    emails = re.findall(r'[\w\.-]+@[\w\.-]+', text)
    return emails

# Example usage:
html_content = """
<html>
<head><title>Sample HTML</title></head>
<body>
<p>This is a sample HTML document with email addresses such as example@email.com and another@example.com.</p>
<p>Contact us at info@company.com for more information.</p>
</body>
</html>
"""

emails_found = extract_emails_from_html(html_content)
print("Email addresses found:", emails_found)
```
