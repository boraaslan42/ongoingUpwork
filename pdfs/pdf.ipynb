```python
import fitz  # PyMuPDF

def extract_text_with_color(pdf_path):
    doc = fitz.open(pdf_path)
    text_with_color = []

    for page in doc:
        for text_block in page.get_text("dict")["blocks"]:
            for line in text_block["lines"]:
                for span in line["spans"]:
                    text = span["text"]
                    color = span["color"]
                    # Check if the color is \uf0a7
                    if color == "\uf0a7":
                        text_with_color.append(text)

    return text_with_color

# Example usage:
pdf_path = "path/to/your/pdf_file.pdf"
text_with_color = extract_text_with_color(pdf_path)
print(text_with_color)
```

```python
from pdf2image import convert_from_path
import pytesseract
from bs4 import BeautifulSoup

def pdf_to_html(pdf_path):
    # Convert PDF pages to images
    images = convert_from_path(pdf_path)
    
    # Perform OCR on each image and extract text
    text = ""
    for img in images:
        text += pytesseract.image_to_string(img)
    
    # Generate HTML markup
    html = "<!DOCTYPE html><html><head><title>PDF to HTML</title></head><body>"
    html += "<div>"
    html += text
    html += "</div></body></html>"
    
    return html

# Example usage
pdf_path = "example.pdf"
html_content = pdf_to_html(pdf_path)

# Write HTML content to a file
with open("output.html", "w", encoding="utf-8") as f:
    f.write(html_content)
```

```python
import PdfReader

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PdfReader.PdfFileReader(file)
        num_pages = reader.numPages
        for page_num in range(num_pages):
            page = reader.getPage(page_num)
            text += page.extractText()
    return text

# Example usage:
pdf_path = pdfList[0][0]+".pdf"
pdf_text = extract_text_from_pdf(pdf_path)
print(pdf_text)
```
